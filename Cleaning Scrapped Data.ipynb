# Cleaning Data i got from WebScrapping [FBREF.com](https://fbref.com/en/comps/9/Premier-League-Stats)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\Owner\OneDrive\Desktop\Excel\PL_12-07-2023.csv', index_col=['Rk'])
```

```python
df.index = np.arange(1,21)
df.index.name = 'Rk'

df
```

```python
df.shape
```

```python
df.size
```

```python
df.info()
```

```python
df = df.drop(columns=['Notes'])

df.describe()
```

```python
df['Ws_in_last_five'] = df['Last 5'].str.count('W')
df['Ls_in_last_five'] = df['Last 5'].str.count('L')
df['Ds_in_last_five'] = df['Last 5'].str.count('D')

df
```

```python
df = df.drop(columns=['Last 5'])

df
```

```python
df['Points_from_last_5'] = (df['Ws_in_last_five']*3) + (df['Ds_in_last_five']*1)

df
```

```python
df['Top Team Scorer'] = df['Top Team Scorer'].str.replace('...', '')
```

```python
df[['Top Team Scorer(s)', 'Top Scorer Total']] = df['Top Team Scorer'].str.split(' - ', expand=True)
df['Top Scorer Total'] = df['Top Scorer Total'].astype(int)
   
df
```

```python
df.info()
```

```python
df = df.drop(columns=['Top Team Scorer'])

df
```

```python
df['Attendance'] = df['Attendance'].str.replace(',', '').astype(int)

df
```
